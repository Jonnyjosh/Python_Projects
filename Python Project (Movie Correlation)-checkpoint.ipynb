```python
# Importing libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

# To adjust the configuration of the plots 
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Reading-in the data
df=pd.read_csv(r"C:\Users\USER\Documents\movies1.csv")
```

```python
# Checking for missing data by looping through the data
for col in df.columns:
    percentage_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percentage_missing))
```

```python
# Changing column(budget and gross) data type

df['budget'] = df['budget'].fillna(0)

df['gross'] = df['gross'].fillna(0)
```

```python
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Creating the (correct year) from the (released) column

df['correctyear'] = df['year'].astype(str).str[0:4]

df
```

```python
# Sorting, by placing highest grossing movies at the top

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# To view all the data on the screen

pd.set_option('display.max_rows', None)
```

```python
  # Removing every duplicates
    
df.drop_duplicates()
```

```python
# Creating a scatter plot that shows the relationship between budget and gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for film')

plt.show()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at company
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
            
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matric for numeric features')

plt.xlabel('movie features')

plt.ylabel('movie features')

plt.show()
```

```python
# Correlation unstacking

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Pairing up

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Pairs with high correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
